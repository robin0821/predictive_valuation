```python
import requests
import time
import numpy as np
import pandas as pd
import csv
import bs4 as BeautifulSoup
import lxml
import re
```

```python
dataset = pd.read_csv("Melbourne_Suburbs_bak.csv", header=None)
```

```python
dataset.head()
```

```python
postcode = dataset.iloc[:,1].values
```

```python
postcode[0]
```

```python
filename = 'melbourne.csv'

# This function write the elements into csv file.
def write_result(price, addr, p_type, bedroom, baths, cars, agent, date, post):
    result = price + addr + str(post) + " is saved!"
    item = [price, addr, p_type, bedroom, baths, cars, agent, date, post]
    csvfile = open(filename, 'a')
    writer = csv.writer(csvfile, dialect = 'excel')
    writer.writerow(item)
    print(result)
    csvfile.close()
```

```python
url0 = r"https://www.realestate.com.au/sold/in-"
```

```python
hds=[{'User-Agent':'Mozilla/5.0 (Windows; U; Windows NT 6.1; en-US; rv:1.9.1.6) Gecko/20091201 Firefox/3.5.6'},
     {'User-Agent':'Mozilla/5.0 (Windows NT 6.2) AppleWebKit/535.11 (KHTML, like Gecko) Chrome/17.0.963.12 Safari/535.11'},
     {'User-Agent': 'Mozilla/5.0 (compatible; MSIE 10.0; Windows NT 6.2; Trident/6.0)'}] 

for post in postcode:
    page_num = 51
    for i in range(1, page_num):
        url = url0 + str(post) + "/list-" + str(i)
        try:
            r = requests.get(url, headers=hds[i%len(hds)])
        except(requests.exceptions.ConnectionError, requests.exceptions.HTTPError, requests.exceptions.ConnectTimeout):
            print("Connection Error, Quit")
        soup = BeautifulSoup.BeautifulSoup(r.content, "lxml")

# If result is less than 1000, try to reset for loop iterations
        if i == 1:
            result_text = soup.find_all("div", {"class": "results-set-header__summary"})[0].text
            result_num = int(re.findall('\d+', result_text)[-1])
            if result_num < 1000:
                page_num = round(result_num/20)

# Else continue means jump out of the for loop. Hence, be careful.
# Main part for extraction
        raw = soup.find_all("article")
        for j in range(0, len(raw)):
            data = raw[j].find("div", {"class": "residential-card__content-wrapper"})
            try:
                price = data.find("span", {"class": "property-price"}).text
                if len(price.strip()) == 0 or price == "Contact agent":
                    break
            except:
                price = "n.a"
            try:
                addr = data.find("a", {"class": "residential-card__info-text"}).text
            except:
                addr = "n.a"
            try:
                p_type = data.find("span", {"class": "residential-card__property-type"}).text
            except:
                p_type = "n.a"
            try:
                bedroom = data.find("span", {"class": "general-features__icon general-features__beds"}).text
            except:
                bedroom = "n.a"
            try:
                baths = data.find("span", {"class": "general-features__icon general-features__baths"}).text
            except:
                baths = "n.a"
            try:
                cars = data.find("span", {"class": "general-features__icon general-features__cars"}).text
            except:
                cars = "n.a"
            try:
                agent = data.find("p", {"class": "agent__name"}).text
            except:
                agent = "n.a"
            try:
                date = data.find("span", {"class": "residential-card__with-comma"}).text
            except:
                date = "n.a"
            write_result(price, addr, p_type, bedroom, baths, cars, agent, date, post)
            
        time.sleep(np.random.rand()*5)
        if len(price.strip()) == 0 or price == "Contact agent":
                break
```
